```python
# Import modules to be used
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
# read in the data
df = pd.read_csv("Downloads/electrical-energy-sales.csv")
df.head(5)
```

```python
#getting all the column details
df.info()
```

```python
# no of values for each state based on year
df["State"].value_counts()
```

### Data Cleaning:

```python
# In the file 'Jharkhand' value has a trailing $ for a value
# Replace it for data consistency
df["State"].replace("Jharkhand \$","Jharkhand" ,regex = True, inplace = True)
```

```python
df["State"].value_counts()
```

```python
# Display values only for a state
State = df[df.State == "Telangana"]
State
```

```python
# count of states
df_states = list(df["State"].unique())
len(df_states)
```

#### Creating a new dataframe by grouping data based on states

```python
# find the sum of domestic consumption in each state
#create two lists for states and their domestic sum
# create a dataframe and print it
lstStates = []
lstTotalDomestic = []
for state in df_states :
    StateData = df[df.State == state]
    StateDomestic = StateData['Domestic'].sum()
    lstStates.append(state)
    lstTotalDomestic.append(StateDomestic)
SeriesStates = pd.Series(lstStates)
SeriesDomestic = pd.Series(lstTotalDomestic)
frame = { 'State' : SeriesStates, 'Domestic' : SeriesDomestic}
ResultDomestic = pd.DataFrame(frame)
print(ResultDomestic)
```

```python
# print all the columns
columns = list[df.columns]
columns
```

```python
# add the Industrial(Low & Med. Voltage) and Industrial(High Voltage)
df["Industrial"] = df["Industrial (Low & Med. Voltage)" ]+ df["Industrial (High Voltage)"]
df
```

```python
# find the total consumption for Commerical,Industrial and Agriculture purpose for the years 
# add it to the dataframe created ResultDomestic
lstStates = []
lstTotalCommercial = []
lstTotalAgriculture = []
lstTotalIndustrial = []
lstStateTotal = []
for state in df_states :
    StateData = df[df.State == state]
    StateCommercial = StateData['Commercial'].sum()
    StateAgriculture = StateData['Agriculture'].sum()
    StateIndustrial= StateData['Industrial'].sum()
    StateTotal = StateData['Total Energy Sold'].sum()
    lstTotalCommercial.append(StateCommercial)
    lstTotalAgriculture.append(StateAgriculture)
    lstTotalIndustrial.append(StateIndustrial)
    lstStateTotal.append(StateTotal)
SeriesCommercial = pd.Series(lstTotalCommercial)
SeriesAgriculture = pd.Series(lstTotalAgriculture)
SeriesIndustrial = pd.Series(lstTotalIndustrial)
SeriesTotal = pd.Series(lstStateTotal)
ResultCommercial = pd.concat([ResultDomestic, SeriesCommercial.rename('Commercial'),], axis=1)
ResultAgri = pd.concat([ResultCommercial, SeriesAgriculture.rename('Agriculture'),], axis=1)
ResultInd = pd.concat([ResultAgri, SeriesIndustrial.rename('Industrial'),], axis=1)
Result = pd.concat([ResultInd,SeriesTotal.rename('Total Consumption'),], axis = 1)
Result
```

```python
# sort the values based on Total comsumption and plot a bar chart for energy consumption of the states 
Result.sort_values("Total Consumption", ascending = False).plot(x = "State", y = ["Total Consumption"], kind="bar", title = "Energy Consumption in India Statewise", figsize =(15,20))
```

```python
# sort the values based on Domestic comsumption and plot a bar chart for Domestic energy consumption of the states 
Result.sort_values("Domestic", ascending = False).plot(x = "State", y = ["Domestic"], kind="bar", title = "Domestic Energy Consumption in India Statewise", figsize =(15,20))
```

```python
# sort the values based on Commercial comsumption and plot a bar chart for energy consumption of the states 
Result.sort_values("Commercial", ascending = False).plot(x = "State", y = ["Commercial"], kind="bar", title = "Commercial Energy Consumption in India Statewise", figsize =(15,20))
```

```python
# sort the values based on Agriculture comsumption and plot a bar chart for Agriculture energy consumption of the states 
Result.sort_values("Agriculture", ascending = False).plot(x = "State", y = ["Agriculture"], kind="bar", title = "Agriculture Energy Consumption in India Statewise", figsize =(15,20))
```

```python
# sort the values based on Industrial comsumption and plot a bar chart for Industrial energy consumption of the states 
Result.sort_values("Industrial", ascending = False).plot(x = "State", y = ["Industrial"], kind="bar", title = "Industrial Energy Consumption in India Statewise", figsize =(15,20))
```

* As per the analysis the highest energy consumer(in all the sectors) in India is Maharshtra. 
* Based on Domestic purpose Uttar Pradesh is the highest consumer.
* On Industrial sector Gujarat is the highest consumer of electricity.

#### Energy consumption increase in India over the 5 years

```python
yearwise
```

```python
yearwise.plot(x = 'Year' , y = 'Total Energy Sold', title = 'Increase in energy consumption over the years' ,kind = 'line')
```
